```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../data/WA_Fn-UseC_-Telco-Customer-Churn.csv")

df.head()
```

```python
df.info()
```

```python
print(df['Churn'].value_counts())
```

```python
churn_counts = df['Churn'].value_counts()

churn_counts.plot(kind='bar')

plt.title("Customer Churn Distribution")
plt.xlabel("Churn")
plt.ylabel("Count")

plt.xticks(rotation=0)

plt.show()
```

```python
contract_churn = pd.crosstab(df['Contract'], df['Churn'])

print(contract_churn)
```

```python
contract_churn.plot(kind='bar', figsize=(8,5))

plt.title("Churn by Contract Type")
plt.xlabel("Contract Type")
plt.ylabel("Customer Count")

plt.xticks(rotation=0)

plt.show()
```

```python
df['MonthlyCharges'].plot(kind='hist', bins=30)

plt.title("Monthly Charges Distribution")
plt.xlabel("Monthly Charges")

plt.show()
```

```python
df['tenure'].plot(kind='hist', bins=30)

plt.title("Customer Tenure Distribution")
plt.xlabel("Tenure")

plt.show()
```

```python
Key Insights:

1. Month-to-month contract customers showed higher churn rates.
2. Long-term contract customers were more likely to stay.
3. Customers with shorter tenure had higher churn probability.
4. Monthly charges may influence customer retention behavior.
5. Businesses should improve retention strategies for high-risk customers.
```
